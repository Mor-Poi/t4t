```python
# [Step 3] Retrieve the cross economic data from Eikon Data API
import eikon as ek
ek.set_app_key('8c4f88cc4b2b4f0a829570f06030f4fd1d5ab9dc')
# Retrieve data point on each Economic Indicator
def get_data(indicator):
    fields = ['TR.IndicatorName','TR.IndicatorType','TR.IndicatorSource'
            ,'ECI_ACT_DT' # GMT Date
            ,'ACT_VAL_NS' # Time
            ,'FCAST_PRD' # Period
            ,'ECON_ACT' # Actual
            ,'RTR_POLL' # Reuters Poll
            ,'ECON_PRIOR' # Prior
            ,'ECON_REV' # Revised
            ,"GN_TXT16_2"# min
            ,"GV4_TEXT"#max
            ]
    
    hk_indicators = indicator
    df, e = ek.get_data(hk_indicators, fields)
    # Rename dataframe columns
    df.rename(columns={'ECI_ACT_DT':'GMT Date'
                    ,'ACT_VAL_NS':'Time'
                    ,'FCAST_PRD':'Period'
                    ,'ECON_ACT':'Actual'
                    ,'RTR_POLL':'Reuters Survey'
                    ,'ECON_PRIOR':'Prior'
                    ,'ECON_REV':'Revised'
                    ,"GN_TXT16_2": 'MIN'
            ,"GV4_TEXT":'MAX'
                    },inplace=True)
    return df
```
